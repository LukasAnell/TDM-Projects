```r
console_sales <- video_games %>%
    group_by(console) %>%
    summarize(total_console_sales = sum(total_sales, na.rm = TRUE)) %>%
    arrange(desc(total_console_sales))
head(console_sales)
```

```r
genre_title_counts <- video_games %>%
    group_by(genre) %>%
    summarize(total_titles = n()) %>%
    arrange(desc(total_titles))
tail(genre_title_counts)
```
