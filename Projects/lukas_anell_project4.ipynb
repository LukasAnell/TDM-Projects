```r
myDF$Passenger_Class <- factor(myDF$Pclass, levels = c(1, 2, 3), labels = c("First Class", "Second Class", "Third Class"))
head(myDF$Passenger_Class)
```

```r
table(myDF$Passenger_Class)
```

```r
myDF$Status <- factor(myDF$Survived, levels = c(0, 1), labels = c("Dead", "Alive"))
head(myDF$Status)
```

```r
myDF$Combined <- paste(myDF$Status, myDF$Sex, sep = " | ")
head(myDF[, c("Status", "Sex", "Combined")], 10)
```

```r
combined_counts <- table(myDF$Combined)
barplot(
    combined_counts,
    col = c("red", "blue", "green", "black"),
    main = "Survival per Sex on the Titanic",
    xlab = "Status and Sex",
    ylab = "Number of Passengers",
    border = "black"
)
```

```r
old_ages <- c(61, 62, 63, 64, 65, 66, 70, 70.5, 71, 74, 80)
old <- myDF$Age[!is.na(myDF$Age) & (myDF$Age %in% old_ages)]
head(old)
```

```r
my_vec1 <- myDF$Age[!is.na(myDF$Age) & (myDF$Age <= 10)]
head(my_vec1)
```

```r
my_vec2 <- myDF$Age[!is.na(myDF$Age) & (myDF$Age > 10 & myDF$Age <= 60)]
head(my_vec2)
```

```r
people <- c(my_vec1, my_vec2)
summary(people)
```

```r
children <- people[people <= 20]
young <- people[people > 20 & people <= 40]
adult <- people[people > 40 & people <= 60]

summary(children)
summary(young)
summary(adult)
```

```r
myDF$AgeGroup[!is.na(myDF$Age) & myDF$Age %in% children] <- "children"
myDF$AgeGroup[!is.na(myDF$Age) & myDF$Age %in% young] <- "young"
myDF$AgeGroup[!is.na(myDF$Age) & myDF$Age %in% adult] <- "adult"
myDF$AgeGroup[!is.na(myDF$Age) & myDF$Age %in% old] <- "old"
table(myDF$AgeGroup)
```

In the instructions, it said to use a table of each range of children, young, and adult to see if they contain the correct values but I prefer summary() because it gives you the min and the max.

4.2: I think splitting up the ages into children, young, adult, and old makes it a lot easier to see how each age demographic was affected in the accident.

4.3: I think I would keep children as is, but change adult to be 30 and above. All of the other categories are fine in my opinion, but I think you qualify more as an adult than as a young person at 30.
